### Matplotlib/Seaborn


* Нарисуйте график sin(x), подпишите оси, добавьте сетку
* Нарисуйте график sin(x) и cos(x) на одном subplot'е и на разных, задайте цвета, которые вам больше нравятся

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
lag = 0.1
x = np.arange(0.0, 10, lag)
y = np.sin(x)

fig = plt.figure()
plt.plot(x, y, color="red")
plt.grid()

plt.title('Sin(x)',
          fontsize=14,
          fontweight="bold",
          color="#808080",
          family="serif")

plt.xlabel('Значение X',
           fontsize=12,
           color="#808080",
           family="serif")
plt.ylabel('Значения Y',
          fontsize=12,
          color="#808080",
          family="serif");
```

```python
lag = 0.1
x = np.arange(0.0, 6.4, lag)
y1 = np.sin(x)
y2 = np.cos(x)

fig = plt.figure()
plt.plot(x, y1, color="red")
plt.plot(x, y2, color="blue")
plt.grid()

plt.title('Sin(x) & Cos(x)',
          fontsize=14,
          fontweight="bold",
          color="#808080",
          family="serif")

plt.xlabel('Значение X',
           fontsize=12,
           color="#808080",
           family="serif")
plt.ylabel('Значения Y',
          fontsize=12,
          color="#808080",
          family="serif");
```

```python
plt.figure(figsize=(12, 7))

plt.subplot(2, 2, 1)
plt.plot(x, y1, color="red")
plt.grid()

plt.title('Sin(x)',
          fontsize=14,
          fontweight="bold",
          color="#808080",
          family="serif")

plt.xlabel('Значение X',
           fontsize=12,
           color="#808080",
           family="serif")

plt.ylabel('Значения Y',
          fontsize=12,
          color="#808080",
          family="serif");

plt.subplot(2, 2, 2)
plt.plot(x, y2, color="blue")
plt.grid()

plt.title('Cos(x)',
          fontsize=14,
          fontweight="bold",
          color="#808080",
          family="serif")

plt.xlabel('Значение X',
           fontsize=12,
           color="#808080",
           family="serif")

plt.ylabel('Значения Y',
          fontsize=12,
          color="#808080",
          family="serif")

plt.subplots_adjust(wspace = 0.3, hspace = 0.3)
```
